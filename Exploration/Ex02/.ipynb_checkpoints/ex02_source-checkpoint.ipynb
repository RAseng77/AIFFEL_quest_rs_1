```python
import cv2
import numpy as np
import torch
import torchvision.transforms as T
from torchvision.models.segmentation import deeplabv3_resnet101
import matplotlib.pyplot as plt

print(cv2.__version__)
```

```python
#pascalvoc 데이터의 라벨종류
LABEL_NAMES = [
    'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
    'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
    'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
]
len(LABEL_NAMES)
```

## Step 1. 인물모드 직접 구현

## (1) 인물 사진

```python
# 준비한 이미지 파일의 경로를 이용하여, 이미지 파일을 읽음
# cv2.imread(경로): 경로에 해당하는 이미지 파일을 읽어서 변수에 저장

person_img_path = "./images/ts_selfie_sgl.jpg"
person_img_orig = cv2.imread(person_img_path)

print(f"이미지 크기: {person_img_orig.shape}")

# cv2.cvtColor(입력 이미지, 색상 변환 코드): 입력 이미지의 색상 채널을 변경
# cv2.COLOR_BGR2RGB: 이미지 색상 채널을 변경 (BGR 형식을 RGB 형식으로 변경)
# plt.imshow(): 저장된 데이터를 이미지의 형식으로 표시, 입력은 RGB(A) 데이터 혹은 2D 스칼라 데이터
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html
# plt.show(): 현재 열려있는 모든 figure를 표시 (여기서 figure는 이미지, 그래프 등)
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html
plt.imshow(cv2.cvtColor(person_img_orig, cv2.COLOR_BGR2RGB))
plt.show()
```

```python

def sematic_segmentation(model, img_orig):

    transform = T.Compose([
        T.ToPILImage(),
        T.Resize((520, 520)),  # 모델 입력 크기 (고정)
        T.ToTensor(),
    ])
    
    input_tensor = transform(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)).unsqueeze(0)
    
    
    with torch.no_grad():
        output = model(input_tensor)["out"][0]
        output_predictions = output.argmax(0).byte().cpu().numpy()
    
    # 원본 크기로 Resize
    output_predictions_resized = cv2.resize(output_predictions, (img_orig.shape[1], img_orig.shape[0]), interpolation=cv2.INTER_NEAREST)

    unique_classes = np.unique(output_predictions_resized)
    print(f"예측된 클래스 ID: {unique_classes}")
    
    for class_id in unique_classes:
        print("ID: ", class_id," ", LABEL_NAMES[class_id])

    plt.imshow(output_predictions_resized, cmap="jet", alpha=0.7)
    plt.title("Segmentation Mask (Resized)")
    plt.show()

    return output_predictions_resized
```

```python
model = deeplabv3_resnet101(pretrained=True).eval()
output_predictions_resized = sematic_segmentation(model, person_img_orig)   
```

```python
# k는 blur의 정도를 조절하는 필터 크기
def blur_cut_img(output_predictions_resized, img_orig, class_id, k):
    # output의 픽셀 별로 예측된 class가 관심 객체라면 1(True), 다르다면 0(False)이 됩니다
    # 1과 0에 각각 255를 곱하였으므로 관심 객체로 예측된 픽셀은 255, 그렇지 않은 픽셀은 0
    # cmap 값을 변경하면 다른 색상으로 확인이 가능함
    seg_map = (output_predictions_resized == class_id)  # 클래스 ID 15 (사람)
    img_mask = seg_map.astype(np.uint8) * 255  # 255 값으로 변환
    plt.imshow(img_mask, cmap='gray')  # 흑백으로 표시
    plt.title("Gray Mask")
    plt.show()
    
    img_orig_blur = cv2.blur(img_orig, (k, k))
    plt.imshow(cv2.cvtColor(img_orig_blur, cv2.COLOR_BGR2RGB))
    plt.title("Blur Image")
    plt.show()

    img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
    # cv2.bitwise_not(): 이미지가 반전됩니다. 배경이 0 객체가 255 였으나
    # 연산을 하고 나면 배경은 255 객체는 0입니다.
    img_bg_mask = cv2.bitwise_not(img_mask_color)
    
    # cv2.bitwise_and()을 사용하면 배경만 있는 영상을 얻을 수 있습니다.
    # 0과 어떤 수를 bitwise_and 연산을 해도 0이 되기 때문에
    # 관심 객체가 0인 경우에는 객체가 있던 모든 픽셀이 0이 됩니다. 결국 객체가 사라지고 배경만 남아요!
    img_bg_blur = cv2.bitwise_and(img_orig_blur, img_bg_mask)
    plt.imshow(cv2.cvtColor(img_bg_blur, cv2.COLOR_BGR2RGB))
    plt.title("Cut Object")
    plt.show()

    # np.where(조건, 참일때, 거짓일때)
    # 세그멘테이션 마스크가 255인 부분만 원본 이미지 값을 가지고 오고
    # 아닌 영역은 블러된 이미지 값을 사용합니다.
    img_concat = np.where(img_mask_color==255, img_orig, img_bg_blur)

    return img_concat
```

```python
img_concat = blur_cut_img(output_predictions_resized, class_id=15, k=30)

# 원본 이미지와 인물사진 모드 이미지 비교
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(person_img_orig, cv2.COLOR_BGR2RGB))
plt.gca().set_title('Original Image' )

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
plt.gca().set_title('Result Image' )

plt.show()
```

## (3) 고양이 + 크로마키

```python
cat_img_path = "./images/cat.jpg"  # 고양이 이미지를 준비해주세요

# 출처 http://365dailyknowledge.files.wordpress.com/2012/01/sahara_desert.jpg
sand_img_path = "./images/desert.JPG"  # 사막 배경 이미지를 준비해주세요

cat_img = cv2.imread(cat_img_path)
sand_img = cv2.imread(sand_img_path)

cat_img = cv2.cvtColor(cat_img, cv2.COLOR_BGR2RGB)
sand_img = cv2.cvtColor(sand_img, cv2.COLOR_BGR2RGB)

print(f"원본 고양이 이미지 크기: {cat_img.shape}")
print(f"배경(사막) 이미지 크기: {sand_img.shape}")

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cat_img)

plt.subplot(1, 2, 2)
plt.imshow(sand_img)

plt.show()
```

```python
model = deeplabv3_resnet101(pretrained=True).eval()

transform = T.Compose([
    T.ToPILImage(),
    T.Resize((520, 520)),  # 모델 입력 크기
    T.ToTensor(),
])

input_tensor = transform(cat_img).unsqueeze(0)

with torch.no_grad():
    output = model(input_tensor)["out"][0]
    output_predictions = output.argmax(0).byte().cpu().numpy()

print(f"추론 마스크 크기 (Before Resize): {output_predictions.shape}")
```

```python
# 마스크를 원본 크기로 Resize
output_predictions_resized = cv2.resize(output_predictions, (cat_img.shape[1], cat_img.shape[0]), interpolation=cv2.INTER_NEAREST)

print(f"추론 마스크 크기 (After Resize): {output_predictions_resized.shape}")

unique_classes = np.unique(output_predictions_resized)
print(f"예측된 클래스 ID: {unique_classes}")

for class_id in unique_classes:
    print("ID: ", class_id," ", LABEL_NAMES[class_id])
```

```python
target_class_id = unique_classes[-1]

seg_map = (output_predictions_resized == target_class_id)
img_mask = seg_map.astype(np.uint8) * 255

plt.imshow(img_mask, cmap='gray')
plt.show()
```

```python
# 배경을 고양이 이미지 크기에 맞춤
sand_img_resized = cv2.resize(sand_img, (cat_img.shape[1], cat_img.shape[0]))

# 고양이 부분만 남기고 배경 적용
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)  # 3채널 변환
result_img = np.where(img_mask_color == 255, cat_img, sand_img_resized)  # 마스크 기반 합성

plt.imshow(result_img)
plt.show()
```

# Step 2. 사진에서 문제점 찾기

- segmentation 되는 영역이 완벽하게 나눠지지 않아 원하지 않은 배경이 나오거나 객체의 일부가 가려지는 문제가 있습니다.
- 특히 테두리가 매끄럽지 않을 경우, 이를 테면 동물과 같이 털이 있을 때 모든 털을 세밀하게 영역을 나눠주지 않기에 잘라낼 경우 어색한 이미지가 생성될 수 있습니다.  
![ex02_problem](./images/ex02_problem.JPG) ![ex02_problem2](./images/ex02_problem_2.JPG) 

# Step 3. 해결 방법 제안

- 세밀하게 피사체의 영역을 구분 짓지 못하는 한계를 극복하기 위해 depth estimation 딥러닝 알고리즘을 활용해 볼 수 있다.   
- Depth estimation은 일반 카메라를 통해서도 객체가 카메라로부터 얼마나 먼 거리에 있는지에 따라 정밀하게 구분지을 수 있다.   
- 3D 카메라를 활용해 볼 수도 있지만, 인물모드를 위해서 3D 카메라로 바꾸는 것은 비효율적인 방법이라고 생각한다.   
- 따라서, 소프트웨어 기술로 객체와의 거리를 추론하여 피사체의 실루엣을 정확히 그리는 방법을 선택한다.   
- Depth estimation task에서 우수한 성능을 보이는 Marigoid 알고리즘을 사용해 볼 수 있다.

![depth_estimation](./images/depth_estimation.JPG)

논문 링크: (https://openaccess.thecvf.com/content/CVPR2024/papers/Ke_Repurposing_Diffusion-Based_Image_Generators_for_Monocular_Depth_Estimation_CVPR_2024_paper.pdf)

구체적인 해결 방법은 아래와 같다.

1. semantic segmentation으로 객체들의 영역을 나눈다.
2. 나눠진 영역들 중 주목하고자 하는 피사체 영역 정보를 저장한다.
3. depth estimation 알고리즘을 활용하여 객체들을 세밀하게 구분 짓는다.
4. 구분된 영역들 중 관심 피사체와 동일한 위치와 거리에 있는 영역으로 segmentation 영역을 확장한다.
